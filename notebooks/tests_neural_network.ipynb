```python
# 1. Définir la métrique personnalisée

def earn_metric(predicted_probs, progressions, n_days):
    base = c = 1
    for j in range(n_days):
        index = len(predicted_probs) - n_days + j
        c *= predicted_probs[index] * progressions[index] + (1 - predicted_probs[index])
        base *= progressions[index]
    return c / base
```

```python
from keras import backend as K

def f1_score(y_true, y_pred):
    def recall(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

    def precision(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision

    precision_value = precision(y_true, y_pred)
    recall_value = recall(y_true, y_pred)
    return 2 * ((precision_value * recall_value) / (precision_value + recall_value + K.epsilon()))
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.metrics import F1Score

# Préparation des données
X = features.values
y = target.values.astype(np.float32)

# Définition du modèle LSTM
model = keras.Sequential([
    layers.InputLayer(input_shape=(X.shape[1], 1)),
    
    # Première couche LSTM avec Dropout et régularisation L2
    layers.LSTM(10, return_sequences=True, 
                dropout=0.2,  # Dropout
                recurrent_dropout=0.2,  # Dropout pour les connexions récurrentes
                kernel_regularizer=regularizers.l2(0.05)),  # Régularisation L2
    
    # Deuxième couche LSTM
    layers.LSTM(4, dropout=0.2, recurrent_dropout=0.2, 
                kernel_regularizer=regularizers.l2(0.05)),
    
    # Couche Dense avec régularisation L2
    layers.Dense(1, activation='sigmoid', kernel_regularizer=regularizers.l2(0.05))
])

# Compilation du modèle avec la métrique F1
f1_metric = F1Score(threshold=0.5)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', f1_metric])

# Paramètres pour la validation en walk-forward
start_index = 2000  # Début de la validation en walk-forward
step_size = 7      # Taille du pas pour chaque itération
total_length = len(X)  # Longueur totale des données
f1_scores = []
predicted_probs = []
progressions = []
earn_metrics = []

# Boucle de validation en walk-forward
for i in range(start_index, total_length, step_size):
    # Définition des indices pour l'entraînement et le test
    train_end = i
    test_end = min(i + step_size, total_length)

    # Séparation des données
    X_train_raw, X_test_raw = X[:train_end], X[train_end:test_end]
    y_train, y_test = y[:train_end], y[train_end:test_end]

    # Normalisation des données
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_raw)
    X_test_scaled = scaler.transform(X_test_raw)

    # Reshape pour LSTM
    X_train = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
    X_test = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)

    # Entraînement du modèle
    model.fit(X_train, y_train, epochs=10, batch_size=2000)
    
    # Prédiction sur le sous-ensemble de test
    prediction_prob = model.predict(X_test)

    # Stocker les probabilités prédites
    predicted_probs.extend(prediction_prob.flatten())

    # Récupérer les progressions réelles pour le sous-ensemble de test
    progressions.extend(data_crop.iloc[train_end:test_end]['progression tomorrow'].values + 1)

    # Calculer la métrique pour cette étape
    earn_metric_value = earn_metric(predicted_probs[-step_size:], progressions[-step_size:], step_size)
    earn_metrics.append(earn_metric_value)

    # Évaluation du modèle sur le sous-ensemble de test
    loss, accuracy, f1_score = model.evaluate(X_test, y_test)
    f1_scores.append(f1_score)
```

```python
import matplotlib.pyplot as plt

plt.plot(f1_scores)
plt.title('F1 Score over Time')
plt.xlabel('Split Index')
plt.ylabel('F1 Score')
plt.show()
```
